```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
plt.figure(figsize=(9,3))
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'b', linestyle = '--', linewidth = 5)
plt.xticks([20,40,60,80,100])
plt.yticks([50,100,150,200])

plt.subplot(1,2,2)
plt.plot(x, z, 'r',  linewidth = 5)
plt.xticks([20,40,60,80,100])


plt.show()
```
